## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
#boxData = df[['ConvertedComp']].transpose()
df['ConvertedComp'].plot(kind = 'box')
#boxData.head()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1 = df['ConvertedComp'].describe()['25%']
Q3 = df['ConvertedComp'].describe()['75%']
IQR = Q3-Q1
print(IQR)
```

Find out the upper and lower bounds.


```python
# your code goes here
upper = Q3 + (1.5*IQR)
lower = Q1 - (1.5*IQR)
print('upper: ', upper,
     'lower: ', lower)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
convertData = df['ConvertedComp'].dropna()
convertData[(convertData>upper) | (convertData< lower)].size
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
new_convertData = convertData[(lower<convertData) & (convertData<upper)]
new_convertData
```
